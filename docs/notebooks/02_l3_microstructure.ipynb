# L2 vs L3 microstructure

The same book can be viewed at **L2 — Market-By-Price** (size aggregated per level) or **L3 — Market-By-Order** (every individual order kept). ob-analytics ships both faces for *comparable* concepts so you can see what the aggregate view hides.

This notebook uses the bundled Bitstamp sample; the identical calls work on LOBSTER data (`Pipeline(format=LobsterFormat(), ...)`), which carries native per-order identity.

```python
from ob_analytics import Pipeline, sample_csv_path

result = Pipeline().run(sample_csv_path())
```

## The book at one instant — aggregate vs per-order

On L2 a price level is a single bar. On L3 that same bar is segmented into its constituent orders — one whale and a crowd of small orders are indistinguishable on L2 and obvious on L3.

```python
result.plot("book_snapshot", "L2")
```

```python
result.plot("book_snapshot", "L3")
```

## Order lifecycles and outcomes

L3 tracks each order from placement to its fate. The lifecycle Gantt colours each lifeline by outcome (filled / cancelled / still resting); the order-outcome face (L3-only) plots placement distance from the touch against size, by fate.

```python
result.plot("order_activity", "L3")
```

```python
result.plot("order_outcome")
```

## Cancellation behaviour

Every cancelled order as a log-log density of *age at cancel* vs *distance from the touch* — the fleeting/patient/deep populations separate that a linear scatter would crush against the origin.

```python
result.plot("cancellations", "L3")
```

## The `prepare` escape hatch

When you need arguments `plot_result` doesn't expose, prepare the payload yourself with the public `prepare` namespace and hand it to `plot`. This is exactly what `result.plot` does under the hood.

```python
from ob_analytics.depth import get_spread
from ob_analytics.visualization import plot, prepare

spread = get_spread(result.depth_summary)
plot(
    "depth_heatmap",
    **prepare.price_levels(result.depth, spread, col_bias=0.1),
)
```
